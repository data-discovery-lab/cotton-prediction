```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
print(sns.__version__)
```

```python
data = [[0, 8267.50, 90.93, 70.50, 8.08, 256.19],
       [1, 8141.86, 90.23, 71.48, 8.19, 290.21],
       [2, 7806.82, 88.35, 67.93, 7.78, 274.25],
       [3, 7382.01, 85.92, 66.54, 7.94, 259.42],
       [4, 6991.91, 83.62, 64.19, 7.45, 261.65],
       [5, 7013.53,  83.75, 63.61, 7.55, 254.38]]

df = pd.DataFrame(data, columns=['ngh_size', 'MSE', 'RMSE', 'MAE', 'MAPE', 'Max Error'])
```

```python
print(df)
```

```python
df2 = pd.melt(df, id_vars=['ngh_size'], 
                  var_name="Type", value_name="Value")
```

```python
df2 = df2.rename(columns={'ngh_size': 'Neighbor Size'})
```

```python
print(df2)
```

```python
sns.set(font_scale=1.5)
sns.lmplot(x='Neighbor Size', y='Value', col='Type', hue='Type', data=df2,
           ci=0, palette='muted', size=4, sharey=False, sharex=False, markers='o',
           scatter_kws={'s': 50, 'alpha': 1}) 
```

```python
sns.set_style("whitegrid")
sns.set(rc={"lines.linewidth": 3})
sns.set_style("ticks", {"xtick.major.size": 1, "ytick.major.size": 1})
sns.set_context("poster")
```

```python
grid = sns.FacetGrid(df2, col="Type", hue="Type", sharey=False, sharex=False, 
                     col_wrap=5, size = 4)


grid.map(plt.plot, "Neighbor Size", "Value", marker="o")
grid.fig.tight_layout(w_pad=1)
```
