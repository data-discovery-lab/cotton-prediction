```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
print(sns.__version__)
```

```python
data = [[0, 8411.54, 91.7, 64.9, 7.9, 235.7],
       [1, 6846.65, 82.74, 64.75, 7.5, 195.33],
       [2, 8109.36, 90.05, 65.88, 7.35, 296.9],
       [3, 7867.77, 88.7, 71.28, 8.3, 257.6],
       [4, 8003.54, 89.46, 70.84, 8.09, 213.25],
       [5, 7903.61, 88.90, 70.40, 8.08, 291.19]]

df = pd.DataFrame(data, columns=['ngh_size', 'MSE', 'RMSE', 'MAE', 'MAPE', 'Max Error'])
```

```python
print(df)
```

```python
df2 = pd.melt(df, id_vars=['ngh_size'], 
                  var_name="Type", value_name="Value")
```

```python
print(df2)
```

```python
sns.set(font_scale=1.5)
sns.lmplot(x='ngh_size', y='Value', col='Type', hue='Type', data=df2,
           ci=0, palette='muted', size=4, sharey=False, sharex=False, markers='o',
           scatter_kws={'s': 50, 'alpha': 1}) 
```

```python
grid = sns.FacetGrid(df2, col="Type", hue="Type", sharey=False, sharex=False, 
                     col_wrap=5, size = 4)


grid.map(plt.plot, "ngh_size", "Value", marker="o")
grid.fig.tight_layout(w_pad=1)
```
